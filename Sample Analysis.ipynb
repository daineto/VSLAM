```python
from unified_planning.shortcuts import *
from unified_planning.io import PDDLReader, PDDLWriter
from VSLAM import *

from copy import deepcopy
from itertools import combinations
import random
import os

import matplotlib.pyplot as plt
```

### CHOOSE DOMAIN

```python
domain_name = "blocks" # from blocks, driverlog, miconic or satellite 
```

Generate demonstration with the original and modified domains

```python
get_environment().credits_stream = None
reader = PDDLReader()

original_demonstrations = set()
extended_demonstrations = set()

original_num_examples = [0]
extended_num_examples = [0]

problems = sorted(os.listdir("benchmarks/{}/problems/".format(domain_name)))
for problem_file in problems:
    print(problem_file)

    problem = reader.parse_problem('benchmarks/{}/domain_original.pddl'.format(domain_name), 'benchmarks/{}/problems/{}'.format(domain_name, problem_file))
    new_transitions, _ = generate_transitions_from_problem(problem, generate_failures=False)
    new_pos = lift_transitions(new_transitions, problem.actions)
    original_demonstrations.update(new_pos)
    original_num_examples += [len(original_demonstrations)]

    problem = reader.parse_problem('benchmarks/{}/domain.pddl'.format(domain_name), 'benchmarks/{}/problems/{}'.format(domain_name, problem_file))
    new_transitions, _ = generate_transitions_from_problem(problem, generate_failures=False)
    new_pos = lift_transitions(new_transitions, problem.actions)
    extended_demonstrations.update(new_pos)
    extended_num_examples += [len(extended_demonstrations)]
```

```python
print(original_num_examples)
print(extended_num_examples)
```

Store data

```python
with open("benchmarks/{}/sample_analysis.csv".format(domain_name), "w") as f:
    for i in range(len(original_num_examples)):
        f.write("{}, {}, {}\n".format(i, original_num_examples[i], extended_num_examples[i]))
```
